<h1>언어 모델(Language Model)</h1>
<pre>: 언어 모델(Language Model, LM)은 언어라는 현상을 모델링하고자 단어 시퀀스(또는 문장)에 확률을 할당하는 모델.
언어 모델을 만드는 방법
1. 통계를 이용한 방법
: 이전 단어들이 주어졌을 때 다음 단어를 예측하는 언어 모델
2. 인공 신경망을 이용한 방법
: 주어진 양쪽의 단어들로부터 가운데 비어있는 단어를 예측하는 언어 모델
 - GPT
 - BERT
 
 언어 모델링(Language Modeling) : 주어진 단어들로부터 아직 모르는 단어를 예측하는 작업
 </pre>
<h2>단어 시퀀스의 확률 할당</h2>
<pre>
언어 모델은 확률틀 통해 보다 적절한 문장 판단
a. 기계 번역(Machine Translation)
: P(나는 버스를 탔다) > P(나는 버스를 태운다)
b. 오타 교정(Spell Correction)
: P(달려갔다) > P(잘려갔다)
c. 음성 인식(Speech Recognition)
: P(나는 메롱을 먹는다) < P(나는 메론을 먹는다)
</pre>

B. 다음 단어 등장 확률

<h1>통계적 언어 모델(Statistical Language Model, SLM)</h1>
<h2>조건부 확률</h2>
P(A), P(B)에 대한 관계<br>



조건부 확률의 연쇄 법칙(chain rule) - n개에 대한 일반화<br>



<h2>문장에 대한 확률</h2>
: 각 단어들이 이전 단어가 주어졌을 때 다음 단어로 등장할 확률의 곱.<br>



<h2>카운트 기반의 접근</h2>
이전 단어로부터 다음 단어에 대한 확률<br>
ex> An addorable little boy가 나왔을 때 is가 나올 확률



An adorable little boy가 100번 등장했을 때, is가 30번 등장한 경우 30%
<h2>카운트 기반 접근의 한계 - 희소 문제(Sparsity Problem)</h2>
: 충분한 데이터를 관측하지 못하여 언어를 정확히 모델링하지 못하는 문제<br>
해결하는 방법으로 n-gram이나 스무딩, 백오프와 같은 여러가지 일반화 기법 존재하지만, 근본적인 해결책은 되지 못하였음. 결국 이러한 한계로 인해 언어 모델의 트렌드는 인공 신경망 언어 모델이 됨.<br><br>

<h1>N-gram 언어 모델(N-gram Language Model)</h1>
여전히 카운트에 기반한 통계적 접근을 사용하고 있으므로 SLM의 일종.<br>
다만, 일반적인 언어 모델과는 달리 이전에 등장한 모든 단어를 고려하는 것이 아니라 일부 단어만 고려하는 접근 방법 사용. 그리고 이 때, 일부 단어를 몇 개 보느냐를 결정하는데 이것이 n-gram에서 n이 가지는 의미.
<h2>코퍼스에서 카운트하지 못하는 경우의 감소.</h2>
SLM의 한계는 훈련 코퍼스에 확률을 계산하고 싶은 문장이나 단어가 없을 수 있다.<br>

확률을 계산하고 싶은 문장이 길어질수록 갖고 있는 코퍼스에서 그 문장이 존재하지 않을 가능성이 높음.<br>

참고하는 단어들을 줄이면 카운트를 할 수 있을 가능성을 높일 수 있음.<br>
<h2>N-gram</h2>
: 임의의 개수를 정하기 위한 기준을 위해 사용하는 것.<br>
n이 1일 때 : 유니그램(unigram)
n이 2일 때 : 바이그램(bigram)
n이 3일 때 : 트라이그램(trigram)
